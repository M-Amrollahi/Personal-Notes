```python
import math
import itertools
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
```

```python
n=30
k=1
p=.7
lst_out = []
for k in range(n):
    lst_out.append(math.factorial(n)/(math.factorial(k)*math.factorial(n-k))*(p**k)*(1-p)**(n-k))
```

```python
%matplotlib inline
sns.lineplot(lst_out)
```

```python
from scipy.stats import bernoulli
import seaborn as sns
import numpy as np

df_bern = bernoulli.rvs(size=10000,p=0.6)

mean = np.mean(df_bern)

var = np.var(df_bern)

print("Mean for Bernoulli distribution is {}". format (mean) )
print("Variance for Bernoulli distribution is {}".format(var))

ax = sns.histplot(df_bern,
kde=True,
color='forestgreen')
ax.set(xlabel='Bernouli', ylabel='Frequency')
plt.title("Bernoulli distribution: p = 0.6")
```

```python
from scipy.stats import norm
import seaborn as sns
import numpy as np

df_bern = norm.rvs(size=1000,loc=9.0,scale=3.0)

mean = np.mean(df_bern)

var = np.var(df_bern)
std = np.std(df_bern)

print("Mean for Bernoulli distribution is {}". format (mean) )
print("Variance for Bernoulli distribution is {}".format(var))
print("STD for Bernoulli distribution is {}".format(std))

ax = sns.histplot(df_bern,
kde=True,
color='forestgreen')
ax.set(xlabel='Bernouli', ylabel='Frequency')
plt.title("Bernoulli distribution: p = 0.6")
```

```python
from scipy.stats import gamma
import seaborn as sns
import numpy as np

df_bern = gamma.rvs(size=1000,a=6)

mean = np.mean(df_bern)

var = np.var(df_bern)
std = np.std(df_bern)

print("Mean for Bernoulli distribution is {}". format (mean) )
print("Variance for Bernoulli distribution is {}".format(var))
print("STD for Bernoulli distribution is {}".format(std))

ax = sns.histplot(df_bern,
kde=True,
color='forestgreen')
ax.set(xlabel='Bernouli', ylabel='Frequency')
plt.title("Bernoulli distribution: p = 0.6")
```

## Beta distribution

```python
from scipy import stats
import seaborn as sns

# Parameters are alpha and beta shape parameters
dist = stats.beta(10, 60)

data = dist.rvs(size=300)

sns.histplot(data,kde=True)
```
